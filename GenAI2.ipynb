## Rate Limits:
* what is Rate Limits?
* Limits set by the API to control how much you can use it (requests, tokens) per minute or per day.

**Rate Limits (simple meaning):**
1. RPM → how many requests you can send per minute
2. RPD → how many requests per day
3. TPM → how many tokens (words/parts of text) per minute
4. TPD → how many tokens per day

#### Default Models:
* Predefined AI models for different tasks (text, coding, image, video), each optimized for speed, quality, or complexity.

#### Default Gemini Models (simple):
* gemini-3-flash-preview → general tasks (text + multimodal)
* gemini-3-pro-preview → coding & complex reasoning
* gemini-2.5-flash-lite → fast + low cost tasks

#### Why gemini-2.5-flash-lite?

1. Fast → gives quick responses (good for practice & testing)
2. Low cost → you can run many times without worrying
3. Enough power → handles basic tasks like Q&A, summaries, small code
4. Low latency → less waiting, faster debugging
5. Best for beginners → no need heavy models for simple learning

Simple idea:
* Use flash-lite → for learning & practice
* Use pro → only when task becomes complex

```python
#Interact Gemini LLM -gemini-2.5-flash-lite
response=client.models.generate_content(
    model='gemini-2.5-flash-lite', # Type of model
    contents='what is the advantage of Gemini 2.5-flash-lite over Gemini 3?'# Prompt[USER/HUMAN]
)
print(response.text)
```

**Observation:**
* The model returned a very large response by default.
* To control the output size and format, we can customize the prompt
* by giving clear instructions (e.g., "give short answer", "in 3 points").

```python
#I want short response(short content)
response=client.models.generate_content(
     model='gemini-2.5-flash-lite', # Type of model
    contents='what is the advantage of Gemini 2.5-flash-lite over Gemini3 in 3 sentence ?'# Prompt[USER/HUMAN]
)
print(response.text)
```

#### system prompt (or) system instructions
* Instead of modifying the prompt every time, we can use a system prompt
* to define fixed instructions (like "give short answers" or "be concise").
* This helps maintain consistent output behavior across all requests
* without repeating instructions in each prompt.

####Simple idea:

* User prompt → changes every time
* System prompt → fixed rules (controls model behavior)


```python
# fix the behavior of llm to respond in 2 to 3 sentences - With help of SYSTEM PROMPT
response = client.models.generate_content(
    model='gemini-2.5-flash-lite',
    contents='When was last recently updated  Gemini 2.5 ?',
    config=genai.types.GenerateContentConfig(
        system_instruction='Always respond to the user queries concisely', #System Prompt or System Instructions or System Messages
    )
)
print(response.text)
```

**Observation:**
* After applying the system prompt, the model returned a very short and precise single-line response.
* This shows that system prompts effectively control the length
 and style of the output without modifying each user prompt.

####What are the llms(models) avalible in grouq?

* I selected the "openai/gpt-oss-120b" model because it is a large and powerful model
 with strong reasoning and coding capabilities.
* It is suitable for handling complex tasks and generating high-quality responses,
 making it a good next step after basic model practice.

```python
#Send a request to the LLM and get response
response= client.chat.completions.create(
    messages=[
        {
            "role": "user", #defines who is sending the message (user input)
            "content": "What are the advantages of generative AI?", # the actual question/prompt
        }
    ],
    model="openai/gpt-oss-120b", #selecting the model to generate response
)

response
```

**Observation:**
* The model returned a structured response (Markdown format) instead of plain text.
* It includes headings, tables, and detailed explanations, making the output rich but lengthy.
* The full response object also contains metadata like tokens used, time taken, and model details.
* To get only the useful answer, we should extract the content part instead of printing the entire response.

```python
print(response.choices[0].message.content)
# Extracts and prints only the actual answer text from the model,
# avoiding extra metadata (tokens, time, etc.) in the full response object.
```

```python
from IPython.display import Markdown
```

* Extract only the model's actual answer text and display it in Markdown format,
* avoiding metadata like tokens, response time, and other extra details.

```python
# Displays only the model's answer in formatted Markdown (headings, tables, etc.)
# while avoiding raw output and metadata.
Markdown(response.choices[0].message.content)
```

#### user prompt and system instructions

```python
# Send request to LLM with both user prompt and system instructions
response = client.chat.completions.create(

    messages=[
        {
            "role": "user",  # user input (question)
            "content": "What are the advantages of generative AI?",  # what we ask
        },
        {
            "role": "system",  # system-level instruction (controls behavior)
            "content": "Respond to user queries concisely"  # tells model to give short answers
        }
    ],

    model="openai/gpt-oss-120b",  # selected powerful model for response generation
)

# Display only the model's answer in formatted Markdown (clean output, no metadata)
Markdown(response.choices[0].message.content)
```

**Observation:**
* The model returned a concise and well-structured answer in bullet points.
* It covers key advantages clearly without unnecessary details.
* The system prompt helped reduce the response size and improve readability.
* This shows that adding a system prompt gives more controlled and focused output.
